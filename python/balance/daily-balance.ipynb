```python
import pandas as pd
import matplotlib.pyplot as plt
#import numpy as np
from datetime import datetime
%matplotlib inline


def days_in(start_date, end_date):
    return (datetime.strptime(end_date, '%Y-%m-%d') - datetime.strptime(start_date, '%Y-%m-%d')). days

#df = pd.read_csv('收支平衡.csv')
df = pd.read_csv('wacai-2014-11-2017-06.csv')
df['日期'] = df['日期'].apply(lambda x: x[:10])
dict_balance = {'收入': 1, '支出': -1}
df['收支'] = df['收支'].map(dict_balance)

incomes = df[df['收支'] == 1]['金额']
spent = df[df['收支']== -1]['金额']

total_income = incomes.sum()
total_spent = spent.sum()

days = days_in(df['日期'].min(), df['日期'].max())
total_trans = df.shape[0]
df
# 总天数
# 总笔数
# 总收入
# 日均收入
# 收入笔数
# 每笔平均收入
# 总支出
# 日均支出
# 支出笔数
# 每笔平均支出
# 每日总资产走势图
```

```python
# 总天数
# 总笔数
print('总天数 = ', days)
print('总笔数 = ', total_trans)
```

```python
# 总收入
# 日均收入
# 收入笔数
# 每笔平均收入
print('总收入 = ', total_income)
print('日均收入 = ', total_income / days)
print('收入笔数 = ', incomes.size)
print('每笔平均收入 = ', total_income / incomes.size)
```

```python
# 总支出
# 日均支出
# 支出笔数
# 每笔平均支出
print('总支出 = ', total_spent)
print('日均支出 = ', total_spent / days)
print('支出笔数 = ', spent.size)
print('每笔平均支出 = ', total_spent / spent.size)
```

```python
l1, = plt.plot(spent, label='spent')
l2, = plt.plot(incomes, label='income')
l3, = plt.plot(spent.apply(lambda x: total_spent / spent.size), label='trans avg cost')
l4, = plt.plot(incomes.apply(lambda x: total_income / incomes.size), label='trans avg income')
plt.legend([l1, l2, l3, l4])
```

```python
sorted = df.sort_values(by='日期')
sorted['金额'] = sorted['金额'] * sorted['收支']
sorted = sorted.drop(['类别', '子类别', '收支', '备注'], axis=1)
date = pd.DataFrame(sorted['日期'].unique())

date.loc[:, '金额'] = pd.Series([0 for i in range(date.shape[0])])
for i, d in date.iterrows():
    date.set_value(i, '金额', sorted[sorted['日期'].__eq__(d[0])]['金额'].sum())
trend = pd.Series([0 for i in range(0, date.shape[0])])
trend[0] = date['金额'][0]

for i in range(1, trend.size):
    trend[i] = date['金额'][i] + trend[i - 1]
```

```python
print('+----------')
print('| 总天数 = ', days)
print('| 总笔数 = ', total_trans)
print('+----------')
print('| 总收入 = ', total_income)
print('| 日均收入 = ', total_income / days)
print('| 收入笔数 = ', incomes.size)
print('| 每笔平均收入 = ', total_income / incomes.size)
print('+----------')
print('| 总支出 = ', total_spent)
print('| 日均支出 = ', total_spent / days)
print('| 支出笔数 = ', spent.size)
print('| 每笔平均支出 = ', total_spent / spent.size)
print('+----------')
print('| 赤字 = ', total_income - total_spent)
print('+----------')

# 每日总资产走势图
plt.plot(trend)
plt.grid()
plt.plot([0 for i in range(0, trend.size)])
```
